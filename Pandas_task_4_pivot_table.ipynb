```python
# подивимось розподіл продажів категорій товарів по містах
goods_in_city = pd.pivot_table ( df, index= 'Product line', columns= 'City', values='Invoice ID', aggfunc='count')
goods_in_city
```

```python
# цей же розподіл, але у відсотках по кожній категорії товарів
perc_goods = goods_in_city.div(goods_in_city.sum(axis=0), axis=1) * 100
```

```python
perc_goods.round (2)
```

```python
# знайдемо категорію, яку найбільше і найменше купують в кожному місті
for city in goods_in_city.columns:
    max_category = goods_in_city[city].idxmax()
    min_category = goods_in_city[city].idxmin()
    max_goods = goods_in_city [city].max()
    min_goods = goods_in_city [city].min()
    max_perc = round ((perc_goods [city].max()),2)
    min_perc = round ((perc_goods [city].min()),2)
    print(f"{city}:")
    print(f" Найбільше продажів: {max_category} {max_goods} і це {max_perc} %")
    print(f" Найменше продажів: {min_category} {min_goods}  і це {min_perc} %\n")
```

```python
# подивимось розподіл використання систем оплат по містам
view_of_payment = pd.pivot_table ( df, index= 'City', columns= 'Payment', values='Invoice ID', aggfunc='count')
view_of_payment.loc ['Total'] = view_of_payment.sum ()
view_of_payment
```
