````python
import pandas as pd
import numpy as np

np.random.seed(0)

products = ['apple', 'banana', 'cherry', 'dates', 'elderberry']
categories = ['fruit', 'fruit', 'fruit', 'fruit', 'fruit']

N_ROWS = 10**6
df = pd.DataFrame({
    'product_name': np.random.choice(products, N_ROWS),
    'product_category': np.random.choice(categories, N_ROWS),
    'product_price': np.random.uniform(1, 10, N_ROWS),
    'transaction_date': pd.date_range(start='1/1/2000', periods=N_ROWS, freq='h')
})

df['product_name'] = df['product_name'].astype('category')
df['product_category'] = df['product_category'].astype('category')

new_memory = df.memory_usage(index=False).sum()
new_memory
````

````python
# відсоток економії пам'яті
print(f"Маємо зниження витрат памʼяті на {round((1-new_memory/first_memory)*100,2)}%")
````

````python
def CalculateVariance(nl):
    sum_list = 0
    for n in nl:
        sum_list = sum_list + n
    mean = sum_list / len(nl)
    sum_squares = 0
    for n in nl:
        sum_squares = sum_squares + n**2
    mean_squares = sum_squares / len(nl)
    return mean_squares
````

````python
input_list = [75, 80, 85, 90, 95]
````

````python
#рефакторінг коду
def calculate_variance(input_list):
    """Обчислює дисперсію для списку чисел."""
    mean = sum(input_list) / len(input_list)
    mean_squares = sum(n ** 2 for n in input_list) / len(input_list)
    variance = mean_squares - mean ** 2
    return variance
````

````python
print("Дисперсія:", calculate_variance(input_list))
````
